```python
# ------------------------------------------------------------------
# 1. 配置与克隆仓库
# ------------------------------------------------------------------
import os
import sys

# [重要] 请在此处替换为你的 GitHub 仓库地址
GITHUB_REPO_URL = "https://github.com/tbuliHe/assignment2-systems"

def setup_environment():
    # 检测是否为 Colab 环境
    is_colab = "google.colab" in sys.modules
    
    if is_colab:
        print("检测到 Google Colab 环境。")
        repo_name = os.path.basename(GITHUB_REPO_URL).replace(".git", "")
        
        # 如果目录不存在，则克隆
        if not os.path.exists(repo_name):
            print(f"正在克隆仓库: {GITHUB_REPO_URL} ...")
            !git clone $GITHUB_REPO_URL
        else:
            print(f"仓库目录 '{repo_name}' 已存在，尝试拉取最新代码...")
            %cd $repo_name
            !git pull
            %cd ..
        
        # 进入项目目录
        if os.path.basename(os.getcwd()) != repo_name:
            os.chdir(repo_name)
            print(f"已切换工作目录至: {os.getcwd()}")
    else:
        print("非 Colab 环境，假设代码已在本地，跳过克隆步骤。")
        print(f"当前工作目录: {os.getcwd()}")

setup_environment()
```
